```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print(sentences)
```

```python
sentences = sent_tokenize(text.replace('\n',' '))
print(sentences)
```

```python
from nltk.tokenize import word_tokenize
words = word_tokenize(text.replace('\n',' '))
print(words)
```

```python
list1 = ['a1','b1','a1']
set1 = set(list1)
print(set1)
```

```python
import string
print("Punctuation marks:", string.punctuation)

clean_words = [word for word in words if word not in set(string.punctuation)]
print(clean_words)
```

```python
characters_to_remove = ["''",'``', "RT", "https", "'s", "'", "--", "'re", "s", '’']
clean_words = [word for word in clean_words if word not in set(characters_to_remove)]

print(clean_words)
```

```python
from nltk.corpus import stopwords
english_stops = set(stopwords.words('english'))

cleanest_words = [word for word in clean_words if word not in english_stops]
print(cleanest_words)
```

```python
clean_tweets=" ".join(str(clean_tweet) for clean_tweet in clean_words)
print(clean_tweets)
```

```python
print(type(clean_tweets))
```

```python
import re


emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
print(emoji_pattern.sub(r'', clean_tweets)) # no emoji
```

```python
import nltk
from nltk.probability import FreqDist

fdist1 = FreqDist(nltk.tokenize.word_tokenize(clean_tweets))

fdist1.most_common(50)
```

```python
# wordcloud package can be used to generate wordclouds
from wordcloud import WordCloud

# patplotlib is a library for plotting graphs
import matplotlib.pyplot as plt

# this setting will display the output inline
%matplotlib inline
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(clean_tweets)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
