```python
#DUMMY CODE FEAIURES HERE:


dflabel= df1
dflabel.vin_number=le.fit_transform(dflabel.vin_number)
dflabel.engine_model=le.fit_transform(dflabel.engine_model)
dflabel.external_color=le.fit_transform(dflabel.external_color)
dflabel.internal_color=le.fit_transform(dflabel.internal_color)
dflabel.engine_type=le.fit_transform(dflabel.engine_type)
dflabel.drive_type=le.fit_transform(dflabel.drive_type)
dflabel.transmission=le.fit_transform(dflabel.transmission)
```

```python
#ADD FEATURES HERE----->
car_model = ols("""price ~ transmission 
                                            + mileage 
                                            + age
                                            + external_color 
                                            + internal_color
                                            + age
                                            + engine_model
                                            + engine_type
                                            + drive_type
                                            + heated_seats_Options
                                            + blind_spot_det
                                            + driver_assist
                                            + harman_kardon_surround_sound
                                            + sport_package
                                            + nav_system
                                            + premium_package
                                            + luxury_package
                                            + parking_assist
                                            + cold_weather_package
                                            + lighting_package
                                            + tech_package
                                            + convenience_package""", data=dflabel).fit()
# summarize our model
car_model_summary = car_model.summary()
HTML(car_model_summary.as_html())
```
